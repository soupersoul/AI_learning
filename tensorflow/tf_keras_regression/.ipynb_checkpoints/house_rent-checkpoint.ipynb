```python
X = df.iloc[:, 0:11].values
y = df.iloc[:, -1].values
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelEncoder = LabelEncoder()
X[:, 0] = labelEncoder.fit_transform(X[:, 0])
X[:, 5] = labelEncoder.fit_transform(X[:, 5])
X[:, 6] = labelEncoder.fit_transform(X[:, 6])
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("city", OneHotEncoder(), [0])], remainder = 'passthrough')
X = ct.fit_transform(X)
print(type(X))
print(X[0:20])
```

```python
# method: 1
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr, X, y, cv=5)
scores
print(scores)
print("Accuracy: %0.2f" % scores.mean())
```

```python
# method 2: tensorflow
model = keras.models.Sequential([
    keras.layers.Dense(30, activation='relu', input_shape=X.shape[1:]),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1)
])
model.compile(loss="mean_squared_error", optimizer="sgd")
#dataset = tf.data.Dataset.from_tensor_slices((X, y))

# tensorflow 不适合用KFold
from sklearn.model_selection import RepeatedStratifiedKFold
rkf = RepeatedStratifiedKFold(n_splits=5, n_repeats=2, random_state=11)
for train_indices, val_indices in rkf.split(X, y):
    train_x = X[train_indices]
    val_x = X[val_indices]
    train_y = y[train_indices]
    val_y = y[val_indices]

    model.fit(train_x, train_y, validation_data=(val_x, val_y), epochs=5)
```

```python
from sklearn.model_selection import train_test_split
train_all_x, test_x, train_all_y, test_y = train_test_split(X, y, random_state=7)
train_x, val_x, train_y, val_y = train_test_split(train_all_x, train_all_y, random_state=11)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_x = scaler.fit_transform(train_x)
val_x = scaler.transform(val_x)
test_x = scaler.transform(test_x)
X = scaler.fit_transform(X)
model = keras.models.Sequential([
    keras.layers.Dense(30, activation='relu', input_shape=X.shape[1:]),
    keras.layers.Dense(1)
])
model.compile(loss="mse", optimizer="adam", metrics=["accuracy"])
history=model.fit(train_x, train_y, validation_data=(val_x, val_y), epochs=30)
```

```python
def plot_learning_curves(history):
    pd.DataFrame(history.history).plot(figsize=[8, 5])
    plt.grid(True)
    plt.gca().set_ylim(0, .01)
    plt.show()
    
plot_learning_curves(history)
```

```python
features = []

categorical_columns = ['city','floor','animal','furniture']
numeric_columns=['area','rooms','bathroom','parking spaces','hoa','rent amount','property tax','fire insurance']
for cc in categorical_columns:
    vocab = df[cc].unique()
    features.append(
        tf.feature_column.indicator_column(
            tf.feature_column.categorical_column_with_vocabulary_list(cc, vocab)
        )
    )
    
for nc in numeric_columns:
    features.append(
        tf.feature_column.numeric_column(nc, dtype=tf.float32)
    )
```

```python
def make_dataset(data_df, value_df, epochs=50, shuffle=True, batch_size=64):
    dataset = tf.data.Dataset.from_tensor_slices((dict(data_df), value_df))
    if shuffle:
        dataset = dataset.shuffle(1000)
    dataset = dataset.repeat(epochs).batch(batch_size)
    return dataset
```

```python
from sklearn.model_selection import train_test_split
x_train_all, x_test, y_train_all, y_test = train_test_split(df, y, random_state=7)

x_train, x_valid, y_train, y_valid = train_test_split(x_train_all, y_train_all, random_state=11)
```

```python
print(x_train.shape)
print(x_valid.shape)
print(type(x_train))
```

```python
output_dir = 'house_rent_model'
if not os.path.exists(output_dir):
    os.mkdir(output_dir)
    
estimator = tf.estimator.DNNRegressor(
        feature_columns = features,
        hidden_units = [50, 100, 50],
        model_dir = output_dir)

estimator.train(input_fn= lambda : make_dataset(x_train, y_train),
                max_steps= 10000)
```

```python
estimator.evaluate(
    input_fn = lambda: make_dataset(
        x_valid, y_valid, epochs=1, shuffle=False
    )
)
```

```python
class WideDeepModel(keras.models.Model):
    def __init__(self):
        super().__init__()
        self.hidden1 = keras.layers.Dense(50, activation='relu')
        self.hidden2 = keras.layers.Dense(50, activation='relu')
        self.output_layer = keras.layers.Dense(1)
        
    def call(self, input):
        hidden1 = self.hidden1(input)
        hidden2 = self.hidden2(hidden1)
        concat = keras.layers.concatenate([input, hidden2])
        output = self.output_layer(concat)
        return output
    
model = WideDeepModel()
model.build(input_shape=(None, train_x[1:]))
model.summary()
model.compile(loss="mse", optimizer='sgd')
print(train_x.shape)
history=model.fit(train_x, train_y, validation_data=(val_x, val_y), epochs=30)
```
